- A vector is a one-dimensional array of numbers, either a row or a column of values.<br/>
- A matrix is a two-dimensional array of numbers arranged in rows and columns. <br/>
- The dot product (or scalar product) is an operation between two vectors of the same length. It multiplies corresponding elements of the vectors and then sums the results.<br/>
- Matrix multiplication is an operation where two matrices, say AA (of size m×nm×n) and BB (of size n×pn×p), are multiplied to produce a new matrix CC (of size m×pm×p).<br/>


```python
# dot product is similar to matrix multiplication and is the standad approch is using np.dot
# parralel processing and fast
#|1*4+2*6 1*5+2*7|
#|3*4+4*6 3*5+5*7|
print(np.dot(matrix_A,matrix_B))
```

### Parabola equation 
<img src = 'https://media.geeksforgeeks.org/wp-content/uploads/20220624194046/par1-300x264.jpg'>

```python
'''
X = [get nmbers from -100 to 100] #list or array 
y = create a function that does following  x**2 = 4ay where  `a` constant or parameter that affects the shape of the graph of the equation influences the "width" or "spread" of the parabola:
y = x**2/4a (a = 1 )
plot x and y check what kind of graph it becomes 
'''
x = array_of_numbers = np.linspace(-100,100)

def y_value(x,a = 1):
    y_array = []
    for number in x:
        conversion = (number**2)/(4*a)
        y_array.append(conversion)
    return y_array

def plot_parabola():
    y =  y_value(x,10)
    fig = plt.figure()
    ax = fig.add_subplot(1, 1, 1)
    ax.spines['left'].set_position('center')
    ax.spines['bottom'].set_position('center')
    ax.spines['right'].set_color('none')
    ax.spines['top'].set_color('none')
    ax.xaxis.set_ticks_position('bottom')
    ax.yaxis.set_ticks_position('left')
    plt.plot(x,y)
    # plt.grid() # create a grid in the plot if we need axis then we write if x = 0 draw straight line 

plot_parabola()
```

```python
'''
Wap that ask for N (number of students to enter record)

First it asks the name of student, Marks in mathematics, marks in science 
Create a data frame at the end and then dump it as data_students.csv

Then in another cell read the csv file from data_students.csv
scatter plot the marks of each students in math and science 
'''
import pandas as pd
def store_record():
    user_input = int(input('Enter the number of Students : '))
    student_dict = {
        'Student' : [],
        'Maths' : [],
        'Science' : []
    }
    for _ in range(user_input):
        student_dict['Student'].append(input('Enter Student Name: '))
        student_dict['Maths'].append((input('Enter Student marks in Maths: ')))
        student_dict['Science'].append(int(input('Enter Student marks in Science: ')))
    
    student_dataframe = pd.DataFrame(student_dict)
    student_dataframe.to_csv('/Users/Uttam/Documents/DSML/Data_student.csv', index = 0)
    print(student_dataframe)

store_record()
```
