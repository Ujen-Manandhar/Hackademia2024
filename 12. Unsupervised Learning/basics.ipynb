#### K-means Clustering 
<p>K-means clustering is an iterative algorithm that tries to partition the dataset into K pre-defined disticnt non - overlaping subgroups (clusters) where each data point belongs to only one group. It tries to make the inter- cluster data points as similar as possible while also keeping the clusters as different (far)as possiblle.</p>

```python
# import necessary libaries
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import random
import seaborn as sns
# if white line then the lib wasn't use 
```

```python
# creating random data 
student = [chr(i + 64) for i in range(1,100)] # generate an ASCII code and convert it into character
math_marks = [random.randrange(1,100) for _ in range(1,100)] # apply random number to the data set 
science_marks = [random.randrange(1,100) for _ in range(1,100)]
print(len(student), len(math_marks), len(science_marks))

# creating data frame 
data_frame = pd.DataFrame(
    {
        'Names': student,
        'Maths': math_marks,
        'Science': science_marks
    }
)
display(data_frame.head(6))
```

```python
# we don't know the end goal.
# from k means clustering we are trying to identify groups by clustering 
# we are trying to group similar kinds of data (or seperate similar group of students), algo in optimize way. 
# this seperation is done by Model.
kmeans = KMeans(n_clusters=4,random_state=42)

# fit the model
kmeans.fit(data_frame[['Maths', 'Science']]) # use kmeans algo

# Add the cluster labels to the Data Frame
data_frame['Cluster'] = kmeans.labels_ #which cluster it belongs to it 
 
display(data_frame.head(random.randrange(1,6)))
# 4d dimension can't be compremended so no scatter plot as 3d is a cube or bred stack
# if more than 3 colums of data then we can't use scatter plot(x,y,z) we can use dimensiolity reduction but model makes sense and visualize it 
# model itself make grouping even if we can't visualize it x1,x2,x3,x4 distance is calculate but graph isn't physically possible  
# clustering and groups are being made trough visualizing through x1,x2,x3,x4 where each is different column AI needs the distance which we can't visualize as it is difficult to compremend; AI needs only distance
# d = sqrt((x2-x1)^2/(y2-y1)^2) the graph dosen't exist in the physical world from physical world each column is a dimension 4 column the 4d
```

```python
# define colors for each cluster (group)
colours = {0: 'red', 1: 'blue', 2: 'green', 3 : 'black'}

# Create the scatter plot 
plt.figure(figsize=(10,6))
for cluster in data_frame['Cluster'].unique():# the unique method puts the 0 value then 1, and 2 sees all the unique value # list of unique values
    subset = data_frame[data_frame['Cluster']== cluster] #subset to identify that makes a data frame of 0 then 1 and then 2 data frame seperate the different data frame 
    plt.scatter(subset['Maths'], subset['Science'], color = colours[cluster], label= f'Cluster {cluster}') # plotting the data on the basis of the data of the cluster and ploting it 


# add labels and title
plt.xlabel('Maths')
plt.ylabel('Science')
plt.title('Scatter plot of maths v/s Science by Cluster')
plt.legend()
plt.show()
```

#### Working on Iris Flower data set

```python
from sklearn.datasets import load_iris

#load the iris dataset
iris = load_iris()
df = pd.DataFrame(data = iris.data, columns=iris.feature_names)
# in this we seperate the group by the model so we don't need the labels

display(df.head(6))
```

```python
#initialize K means with 3 clusters (setosa, versicolor and virginica)
kmeans = KMeans(n_clusters=3, random_state=42)

# fit the model
kmeans.fit(df)

# add the cluster labels to the data frame 
df['Cluster'] = kmeans.labels_
display(df.head(6))

# Map cluster labels to actual species names
species_map = {0: 'Setosa', 1: 'Versicolor', 2: 'Virginica'}
df['Predicted Species'] = df['Cluster'].map(species_map)# .map is a function of pandas 
# It's commonly used to transform or replace values based on a mapping you define.
display(df.head(6))
```

<p>Iris flower set is 4D and as such we can't visualize it but k means is working in the four points; and on the basis of the four point the cluster is done in three groups but in case of visualization its impossible so we are visualizing on the basis of	sepal length (cm) and sepal width (cm). And again  petal length (cm) and petal width (cm). It was solve in 4D but we can't visualize it.</p>

```python
# define a custom palerre with distict color for each sepecies
custom_palette = {'Setosa': 'red', 'Versicolor': 'blue', 'Virginica': 'green'}

#visualize the cluster with actual species name and disticnt colors
plt.figure(figsize=(10,6))
sns.scatterplot(x='sepal length (cm)', y= 'sepal width (cm)', hue='Predicted Species', data=df, palette=custom_palette)
plt.title('K-means Clustering (k=3) with Flower Names')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x='petal length (cm)', y= 'petal width (cm)', hue='Predicted Species', data=df, palette=custom_palette)
plt.title('K-means Clustering (k=3) with Flower Names')
plt.show()
```

#### Principal Component Analysis (PCA)
<p>PCA is a dimensionality reduction technique that is often used to reduce the dimension of large datasets, by transforming the data into a new set of variables, the principal components, which are orthogonal(uncorrelated) and ordered by the amount of orginal variance they retain. Watch a statquest</p>
<p>Every points of the class is sifted to the center using its average; from there an arbritary value is drawn which is a line that defines all the points (like linear regression) this line is one axis, then it is done with another class and a arbritary line is made, this line is perpendicular to the first line(new xy axis is made.) </p>

```python
# for PCA of Iris flower data set the 4D are converted into PCA 1 and PCA 2 and on the basis of PCA 1 and 2 clustering is done ie the class identified.
from sklearn.decomposition import PCA

# asuming df is already defined and has only numerica data except 'Cluster' and 'Predicted species' above
# drop non-numberic colums (like 'Cluster' and 'Predicted Species')for PCA

df_numeric = df.drop(columns=['Cluster', 'Predicted Species'], axis=1)
display(df_numeric.head(6))
```

```python
# Initialize PCA
pca = PCA(n_components=2)

# Fit and tranform the data
pca_result = pca.fit_transform(df_numeric)

# Create a dataframe for the PCA result
df_pca = pd.DataFrame(data = pca_result, columns=['PCA 1', 'PCA 2'])

# create a Dataframe for thr PCA dataframe
df_pca['Species'] = df['Predicted Species']

# visualize the PCA result
plt.figure(figsize=(10,6))
sns.scatterplot(x='PCA 1', y='PCA 2', hue= 'Species', data = df_pca, palette='viridis' )
plt.title('PCA of IRIS data set')
plt.show()
```
